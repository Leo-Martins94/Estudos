```python
#Imports
import time
import numpy as np
import pandas as pd
import matplotlib as mat
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
%matplotlib inline

fontsize = 14
ticklabelsize = 14
```

```python
#Carregando o dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
print(len(df))
df.head()
```

## Plot

### Crie um Histograma de sepal length

```python
exclude = ['Id', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)', 'target']
df.loc[:, df.columns.difference(exclude)].hist()
plt.figure(figsize=(15,10))
plt.show()
```

### Crie um Gráficos de Dispersão (scatter Plot) da variábel sepal length versus npumero da linha, colorido por marcadores da variável target

```python
plt.figure(figsize=(12, 8), dpi=80)
plt.scatter(range(len(df)), df['petal width (cm)'], c=df['target'])
plt.xlabel('Número de Linha', fontsize=fontsize)
plt.ylabel('Sepal length (cm)', fontsize=fontsize)
plt.title('Gráfico de Disperção ds Atributos, colorido por marcadores da classe alvo', fontsize=fontsize)
#plt.title('Scatter plot of fatures, colored by target labels', fontsize=fontsize)
```

### Crie um Scatter Plot de 2 Features (atributos)

```python
plt.figure(figsize=(12, 8), dpi=80)
plt.scatter(df['petal length (cm)'], df['petal width (cm)'], c=df['target'])
plt.xlabel('petal length (cm)', fontsize=fontsize)
plt.ylabel('petal width (cm)', fontsize=fontsize)
```

### Crie um Scatter Matrix das Features (Atributos)
